## 1. Importar Librerías

Primero importamos todas las librerías necesarias para el análisis.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from datetime import datetime
import warnings

warnings.filterwarnings('ignore')

# Configuración de matplotlib para que se vea lindo
plt.style.use('seaborn-v0_8-darkgrid')
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 10

print("✅ Librerías importadas correctamente")
```

## 2. Cargar Datos del Juego

Cargamos el CSV generado por las partidas jugadas.

```python
# Cargar datos
try:
    df = pd.read_csv("../data/game_data.csv")
    print(f"✅ Datos cargados correctamente")
    print(f"📊 Total de turnos registrados: {len(df)}")
    print(f"🎮 Número de partidas: Aproximadamente {len(df)//20}")  # Asumiendo 20 turnos por partida
    print("\n👀 Vista previa de los datos:")
    display(df.head(10))
except FileNotFoundError:
    print("❌ No se encontró el archivo game_data.csv")
    print("🎮 Primero tenés que jugar al menos una partida!")
    print("   Ejecutá: cd ../game && python main.py")
```

## 3. Exploración de Datos

Estadísticas descriptivas de las variables económicas.

```python
# Información general del dataset
print("📋 INFORMACIÓN DEL DATASET")
print("="*60)
print(df.info())
print("\n📊 ESTADÍSTICAS DESCRIPTIVAS")
print("="*60)
display(df.describe().round(2))
```

## 4. Análisis de Decisiones

¿Qué decisiones se tomaron más frecuentemente?

```python
# Conteo de decisiones
print("🎯 DECISIONES MÁS TOMADAS")
print("="*60)
decision_counts = df['decision'].value_counts()
print(decision_counts)

# Gráfico de barras
plt.figure(figsize=(14, 6))
decision_counts.plot(kind='barh', color='steelblue')
plt.title('Frecuencia de Decisiones Tomadas', fontsize=16, fontweight='bold')
plt.xlabel('Cantidad de veces', fontsize=12)
plt.ylabel('Decisión', fontsize=12)
plt.tight_layout()
plt.show()
```

## 5. Evolución de Variables Económicas

Gráficos de cómo evolucionan las variables a lo largo de los turnos.

```python
# Gráfico de evolución de variables principales
fig, axes = plt.subplots(2, 2, figsize=(16, 10))

# PBI
axes[0, 0].plot(df['turno'], df['pbi'], marker='o', color='darkgreen', linewidth=2)
axes[0, 0].set_title('📈 Evolución del PBI', fontsize=14, fontweight='bold')
axes[0, 0].set_xlabel('Turno')
axes[0, 0].set_ylabel('PBI')
axes[0, 0].grid(True, alpha=0.3)
axes[0, 0].axhline(y=100, color='r', linestyle='--', alpha=0.5, label='Valor inicial')
axes[0,  0].legend()

# Felicidad
axes[0, 1].plot(df['turno'], df['felicidad'], marker='o', color='orange', linewidth=2)
axes[0, 1].set_title('😊 Evolución de la Felicidad', fontsize=14, fontweight='bold')
axes[0, 1].set_xlabel('Turno')
axes[0, 1].set_ylabel('Felicidad (%)')
axes[0, 1].grid(True, alpha=0.3)
axes[0, 1].axhline(y=50, color='r', linestyle='--', alpha=0.5, label='Valor inicial')
axes[0, 1].legend()

# Reservas
axes[1, 0].plot(df['turno'], df['reservas'], marker='o', color='blue', linewidth=2)
axes[1, 0].set_title('💰 Evolución de las Reservas', fontsize=14, fontweight='bold')
axes[1, 0].set_xlabel('Turno')
axes[1, 0].set_ylabel('Reservas (M USD)')
axes[1, 0].grid(True, alpha=0.3)
axes[1, 0].axhline(y=20000, color='r', linestyle='--', alpha=0.5, label='Valor inicial')
axes[1, 0].legend()

# Inflación
axes[1, 1].plot(df['turno'], df['inflacion'], marker='o', color='red', linewidth=2)
axes[1, 1].set_title('🔥 Evolución de la Inflación', fontsize=14, fontweight='bold')
axes[1, 1].set_xlabel('Turno')
axes[1, 1].set_ylabel('Inflación (%)')
axes[1, 1].grid(True, alpha=0.3)
axes[1, 1].axhline(y=5, color='g', linestyle='--', alpha=0.5, label='Valor inicial')
axes[1, 1].legend()

plt.tight_layout()
plt.show()
```

```python
# Gráfico adicional: Todas las variables en uno
fig, ax = plt.subplots(figsize=(16, 8))

# Normalizar para poder comparar
df_norm = df.copy()
variables = ['pbi', 'felicidad', 'reservas', 'inflacion', 'desempleo', 'dolar']

for var in variables:
    # Normalizar cada variable a escala 0-100
    min_val = df[var].min()
    max_val = df[var].max()
    if max_val - min_val != 0:
        df_norm[var + '_norm'] = (df[var] - min_val) / (max_val - min_val) * 100
    else:
        df_norm[var + '_norm'] = 50

# Graficar
ax.plot(df_norm['turno'], df_norm['pbi_norm'], marker='o', label='PBI', linewidth=2)
ax.plot(df_norm['turno'], df_norm['felicidad_norm'], marker='s', label='Felicidad', linewidth=2)
ax.plot(df_norm['turno'], df_norm['reservas_norm'], marker='^', label='Reservas', linewidth=2)
ax.plot(df_norm['turno'], 100 - df_norm['inflacion_norm'], marker='v', label='Estabilidad (inv. inflacion)', linewidth=2)
ax.plot(df_norm['turno'], 100 - df_norm['desempleo_norm'], marker='d', label='Empleo (inv. desempleo)', linewidth=2)

ax.set_xlabel('Turno', fontsize=12)
ax.set_ylabel('Valor Normalizado (0-100)', fontsize=12)
ax.set_title('🎯 Dashboard Económico - Todas las Variables Normalizadas', fontsize=16, fontweight='bold')
ax.legend(loc='best', fontsize=10)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 6. Matriz de Correlación

¿Cómo se relacionan las variables entre sí?

```python
# Matriz de correlación
variables_numericas = ['pbi', 'felicidad', 'reservas', 'inflacion', 'desempleo', 'deuda_externa', 'dolar', 'depositos_bancarios']
correlation_matrix = df[variables_numericas].corr()

# Crear heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, 
            square=True, linewidths=1, cbar_kws={"shrink": 0.8}, fmt='.2f')
plt.title('🔥 Matriz de Correlación - Variables Económicas', fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()

print("\n🔍 INTERPRETACIÓN:")
print("="*60)
print("✅ Verde oscuro = Correlación positiva fuerte")
print("❌ Rojo oscuro = Correlación negativa fuerte")
print("⚪ Blanco = Sin correlación")
```

## 7. Impacto Promedio por Decisión

¿Qué decisiones fueron más efectivas?

```python
# Agrupar por decisión y calcular promedios
impact_by_decision = df.groupby('decision')[['pbi', 'felicidad', 'reservas', 'inflacion', 'desempleo']].mean()

print("📊 IMPACTO PROMEDIO DE CADA DECISIÓN")
print("="*60)
display(impact_by_decision.round(2))

# Visualizar impacto
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# PBI promedio por decisión
impact_by_decision['pbi'].sort_values().plot(kind='barh', ax=axes[0, 0], color='darkgreen')
axes[0, 0].set_title('📈 PBI Promedio por Decisión', fontweight='bold')
axes[0, 0].set_xlabel('PBI Promedio')
axes[0, 0].axvline(x=100, color='r', linestyle='--', alpha=0.5)

# Felicidad promedio por decisión
impact_by_decision['felicidad'].sort_values().plot(kind='barh', ax=axes[0, 1], color='orange')
axes[0, 1].set_title('😊 Felicidad Promedio por Decisión', fontweight='bold')
axes[0, 1].set_xlabel('Felicidad Promedio (%)')
axes[0, 1].axvline(x=50, color='r', linestyle='--', alpha=0.5)

# Reservas promedio por decisión
impact_by_decision['reservas'].sort_values().plot(kind='barh', ax=axes[1, 0], color='blue')
axes[1, 0].set_title('💰 Reservas Promedio por Decisión', fontweight='bold')
axes[1, 0].set_xlabel('Reservas Promedio (M USD)')
axes[1, 0].axvline(x=20000, color='r', linestyle='--', alpha=0.5)

# Inflación promedio por decisión
impact_by_decision['inflacion'].sort_values().plot(kind='barh', ax=axes[1, 1], color='red')
axes[1, 1].set_title('🔥 Inflación Promedio por Decisión', fontweight='bold')
axes[1, 1].set_xlabel('Inflación Promedio (%)')
axes[1, 1].axvline(x=5, color='g', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

## 8. Análisis de Eventos Random

¿Cómo afectaron los eventos random al juego?

```python
# Filtrar solo los turnos donde hubo eventos
df_with_events = df[df['event'] != ''].copy()

print(f"🎲 Total de eventos que ocurrieron: {len(df_with_events)}")
print(f"📊 Porcentaje de turnos con eventos: {len(df_with_events)/len(df)*100:.1f}%")
print("\n🎯 EVENTOS MÁS FRECUENTES:")
print("="*60)
print(df_with_events['event'].value_counts())

# Graficar eventos más frecuentes
if len(df_with_events) > 0:
    plt.figure(figsize=(14, 6))
    df_with_events['event'].value_counts().plot(kind='barh', color='purple')
    plt.title('🎲 Frecuencia de Eventos Random', fontsize=16, fontweight='bold')
    plt.xlabel('Cantidad de veces')
    plt.ylabel('Evento')
    plt.tight_layout()
    plt.show()
else:
    print("\n⚠️ No hubo eventos random en esta partida")
```

## 9. Detección de Momentos de Crisis

Identificamos los momentos más críticos del gobierno.

```python
# Definir criterios de crisis
df['en_crisis'] = (
    (df['inflacion'] > 50) | 
    (df['desempleo'] > 25) | 
    (df['felicidad'] < 20) | 
    (df['reservas'] < 5000)
)

crisis_moments = df[df['en_crisis'] == True]

print(f"⚠️ MOMENTOS DE CRISIS DETECTADOS: {len(crisis_moments)}")
print(f"📊 Porcentaje de turnos en crisis: {len(crisis_moments)/len(df)*100:.1f}%")

if len(crisis_moments) > 0:
    print("\n🚨 TURNOS EN CRISIS:")
    print("="*60)
    display(crisis_moments[['turno', 'decision', 'pbi', 'felicidad', 'reservas', 'inflacion', 'desempleo']])
    
    # Visualizar crisis en timeline
    plt.figure(figsize=(16, 6))
    plt.plot(df['turno'], df['felicidad'], marker='o', linewidth=2, label='Felicidad')
    plt.scatter(crisis_moments['turno'], crisis_moments['felicidad'], 
                color='red', s=200, marker='X', label='Momento de Crisis', zorder=5)
    plt.xlabel('Turno', fontsize=12)
    plt.ylabel('Felicidad (%)', fontsize=12)
    plt.title('🚨 Momentos de Crisis en el Timeline', fontsize=16, fontweight='bold')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
else:
    print("\n✅ ¡Felicitaciones! No hubo momentos de crisis extrema.")
```

## 10. Resumen Ejecutivo

Métricas clave del gobierno.

```python
# Resumen ejecutivo
print("=" * 70)
print(" " * 20 + "📋 RESUMEN EJECUTIVO")
print("=" * 70)

# Valores iniciales vs finales
pbi_inicial = df.iloc[0]['pbi']
pbi_final = df.iloc[-1]['pbi']
felicidad_inicial = df.iloc[0]['felicidad']
felicidad_final = df.iloc[-1]['felicidad']
reservas_inicial = df.iloc[0]['reservas']
reservas_final = df.iloc[-1]['reservas']
deuda_inicial = df.iloc[0]['deuda_externa']
deuda_final = df.iloc[-1]['deuda_externa']

print(f"\n🏛️ GESTIÓN GOBIERNO:")
print(f"   Turnos totales: {len(df)}")
print(f"   Decisiones tomadas: {df['decision'].nunique()}")
print(f"   Eventos enfrentados: {len(df_with_events)}")

print(f"\n📊 EVOLUCIÓN ECONÓMICA:")
cambio_pbi = ((pbi_final - pbi_inicial) / pbi_inicial * 100)
emoji_pbi = "📈" if cambio_pbi > 0 else "📉"
print(f"   {emoji_pbi} PBI: {pbi_inicial:.1f} → {pbi_final:.1f} ({cambio_pbi:+.1f}%)")

cambio_deuda = ((deuda_final - deuda_inicial) / deuda_inicial * 100)
emoji_deuda = "⚠️" if cambio_deuda > 0 else "✅"
print(f"   {emoji_deuda} Deuda Externa: ${deuda_inicial:,.0f}M → ${deuda_final:,.0f}M ({cambio_deuda:+.1f}%)")

cambio_reservas = ((reservas_final - reservas_inicial) / reservas_inicial * 100)
emoji_reservas = "💰" if cambio_reservas > 0 else "💸"
print(f"   {emoji_reservas} Reservas: ${reservas_inicial:,.0f}M → ${reservas_final:,.0f}M ({cambio_reservas:+.1f}%)")

print(f"\n😊 SATISFACCIÓN POPULAR:")
cambio_felicidad = felicidad_final - felicidad_inicial
emoji_felicidad = "😊" if cambio_felicidad > 0 else "😡"
print(f"   {emoji_felicidad} Felicidad: {felicidad_inicial:.1f}% → {felicidad_final:.1f}% ({cambio_felicidad:+.1f} puntos)")

print(f"\n🎯 CALIFICACIÓN FINAL:")
# Calcular score basado en métricas finales
score = (pbi_final + felicidad_final + (reservas_final/1000)) / 3
if score > 80:
    calificacion = "⭐⭐⭐⭐⭐ EXCELENTE - Sos mejor que todos los presidentes reales"
elif score > 60:
    calificacion = "⭐⭐⭐⭐ BUENO - El país no explotó"
elif score > 40:
    calificacion = "⭐⭐⭐ REGULAR - Sobreviviste de milagro"
elif score > 20:
    calificacion = "⭐⭐ MALO - La historia te juzgará duramente"
else:
    calificacion = "⭐ DESASTROSO - Peor que el 2001 real"

print(f"   {calificacion}")
print(f"   Score: {score:.1f}/100")

print("\n" + "=" * 70)
```
